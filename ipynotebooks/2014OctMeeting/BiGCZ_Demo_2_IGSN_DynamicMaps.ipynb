```python
# Convert XML to dict structure easily exported to JSON
# Downloaded on 11/29/2014 from 
# http://charlieandrews.tumblr.com/post/52795541182/parsing-xml-to-json-in-python
#from lxml import objectify
#import json
# This follows google's rules for conversion of XML to JSON
def iterNodes(node, parentDict):
    nodeDict = {}
    try:
        nodeDict.update(node.attrib)
    except AttributeError:
        pass
    if node.text != None:
        nodeDict['$t'] = node.text
    
    for i in node.iterchildren():
        childDict = {}
        newDict = {}
        newDict = iterNodes(i, childDict)
        newList = []
        if i.tag in nodeDict:
            try:
                nodeDict[i.tag].append(newDict[i.tag])
            except:
                newList.append(nodeDict[i.tag])
                nodeDict[i.tag] = newList
                nodeDict[i.tag].append(newDict[i.tag])
        else:
            nodeDict.update(newDict)
    tagList = node.tag.split(':')
    namespace = '$'.join(tagList)
    parentDict[namespace] = nodeDict
    return parentDict
```

### Plot IGSN sample on a map, using Folium package
Code adapted from https://github.com/wrobstory/folium/blob/master/examples/marker_color_icon_examples.ipynb

```python
from IPython.display import HTML
import folium
```

```python
def inline_map(map):
    """
    Embeds the HTML source of the map directly into the IPython notebook.
    
    This method will not work if the map depends on any files (json data). Also this uses
    the HTML5 srcdoc attribute, which may not be supported in all browsers.
    """
    map._build_map()
    return HTML('<iframe srcdoc="{srcdoc}" style="width: 100%; height: 510px; border: none"></iframe>'.format(srcdoc=map.HTML.replace('"', '&quot;')))

def embed_map(map, path="m213map.html"):
    """
    Embeds a linked iframe to the map into the IPython notebook.
    
    Note: this method will not capture the source of the map into the notebook.
    This method should work for all maps (as long as they use relative urls).
    """
    map.create_map(path=path)
    return HTML('<iframe src="files/{path}" style="width: 100%; height: 510px; border: none"></iframe>'.format(path=path))
```

### Ingesting geojson (__geo_interface__) point features into a GeoDataFrame

```python
from geojson import Point, Feature, FeatureCollection
```

```python
dfctr = gpd.GeoDataFrame(geometry=df.centroid, crs=df.crs)
```

```python
dfctr.to_json()
```

```python
# Point() takes X, Y (easting, northing)
feature1 = Feature(geometry=Point((-122.5, 47.5)), id=1, properties={"sample_type": "soil"})
feature2 = Feature(geometry=Point((-122, 47)), id=2, properties={"sample_type": "water"})
```

```python
gjfc = FeatureCollection([feature1, feature2])
gjfc
```

```python
gjfc.__geo_interface__
```

```python
# gpd.GeoDataFrame.from_features(features=gjfc) fails
# Looks like features expects a list (iterable) of features, but can't handle a FeatureCollection
# doc says: "Each element must be a feature dictionary or implement the __geo_interface__"
dffromgjson = gpd.GeoDataFrame.from_features(features=gjfc['features'])
# Could also use features=[feature1, feature2]
```

```python
# the points are barely visible here and in the mplleaflet map; are tiny black/blue dots
axgj = dffromgjson.plot()
```
