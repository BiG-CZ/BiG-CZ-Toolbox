### Demo using HIS Central GetSeriesCatalogForBox2, then WOF GetValuesObject
[I originally presented this as a blog post back in 2010.](http://miestero.wordpress.com/2010/07/26/python-waterml-numpy_3/)

```python
from suds.client import Client
import xml.etree.ElementTree as ET
import numpy as NP
import datetime as DT
 
# lambda function to convert ISO datetime string to python datetime object
isodtstr2dt = lambda isodtstr: DT.datetime.strptime(
                     isodtstr, "%Y-%m-%dT%H:%M:%S")

#URL to the HIS Central API
# 5/14/2014. The current, updated WSDL from the Water Data Center (WDC) is:
#HIS_Central_URL = "http://hiscentral.cuahsi.org/webservices/hiscentral.asmx?WSDL"
# The one below is obsoloscent and out of date, 
# but my old code worked fine with and not with the WDC WSDL :()
HIS_Central_URL = "http://water.sdsc.edu/hiscentral/webservices/hiscentral.asmx?WSDL"
```

```python
#Search parameters
bbox1 = (45.590978,-121.871338,45.870888,-119.910278) # Lower Columbia river
bbox2 = (33.84, -81.25, 34.24, -80.84) # South Carolina?
ymin,xmin,ymax,xmax = bbox2
keyword = 'Streamflow' # A higher-level, common vocabulary mapping from individual variable names
start_date = '2000-01-01'
end_date   = '2000-12-31'
 
#query HIS Central for time series
client = Client(HIS_Central_URL)
response = client.service.GetSeriesCatalogForBox2(xmin, xmax, ymin, ymax, 
                                                  keyword, '', start_date, end_date)
# Returns list of dicts holding metadata for each matching series
series_array = response[0]
```

```python
#for each series, call GetValues on the data server
series_location_lst = []
plt.figure(figsize=(15,6), dpi=100)
for series in series_array:
    client = Client(series.ServURL)
    values_obj = client.service.GetValuesObject(series.location,
                       series.VarCode, start_date, end_date)
    tsvalues_obj = values_obj.timeSeries.values
 
    #if more than 10 values, then add to plot
    if hasattr(tsvalues_obj, '_count') and int(tsvalues_obj._count) > 50:
        val_a = NP.asarray([float(val.value)
                           for val in tsvalues_obj.value])
        dt_a  = NP.asarray([val._dateTime
                           for val in tsvalues_obj.value])
        #dt_a  = NP.asarray([isodtstr2dt(val._dateTime)
        #                  for val in tsvalues_obj.value])
 
        series_location_lst.append(series.location)
        plt.plot_date(dt_a, val_a, '-o', markersize=2.5)
plt.title("%s time series for bounding box (%.2f, %.2f, %.2f, %.2f)" % (keyword, ymin,xmin,ymax,xmax))
plt.legend(series_location_lst);
```
