```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

Note that some (or many?) CZOData display files include a single file both the \header (and \doc) and \data blocks. This particular type of file handled here separates them, leaving a simplified data file, but one that seems a bit non-standard (ie, I thought the data block did not have column headers directly, but these had to be interpreted based on column order and the column \header block)

```python
baseurl = "http://czo.stroudcenter.org/data/hydrology"
datafile_tmplt = "%s/CRB_WCC_STAGEFLOW_%d.csv"  # % (baseurl, yearint)
# A sample file:
# http://czo.stroudcenter.org/data/hydrology/CRB_WCC_STAGEFLOW_2009.csv
# Header file:
# http://czo.stroudcenter.org/data/hydrology/CRB_WCC_STAGEFLOW_header.csv
```

```python
# Read each URL (csv file) using the Pandas "read_csv" function, 
# then append the resulting DataFrame object to a temporary list called "data_frames".
years = range(2000, 2013, 1)

data_frames = []
for yr in years:
    csvurl = datafile_tmplt % (baseurl, yr)
    df = pd.read_csv(csvurl, header=0, skipinitialspace=True, skiprows=[1], index_col=0, na_values=[-9999], parse_dates=True)
    data_frames.append(df)
```

```python
# Concatenate all the DataFrames in the "data_frames" list into a single DataFrame
df = pd.concat(data_frames)
df
```

```python
df.index
```

```python
df.index = df.index.tz_localize('EST')
df.index
```

```python
df.dtypes
```

```python
df = df.convert_objects(convert_numeric='force')
df.dtypes
```

```python
plt.figure(figsize=(12,5), dpi=100)
var_discharge = df['Discharge (cfs) from Continuous record']
var_stage     = df['Gage Height (ft) from Continuous record']
ax = var_discharge.plot(title='CRB Hydro Data', style='b', logy=True, ylim=(1,1000), legend=True)
ax.set_ylabel(var_discharge.name, color='b')
ax = var_stage.plot(secondary_y=True, style='g', legend=True)
ax.right_ax.set_ylabel(var_stage.name, color='g')
```
